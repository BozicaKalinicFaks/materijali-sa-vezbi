```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def kernel(x1, x2, bandwidth):
    return np.exp(-0.5*((x1-x2)**2) / (bandwidth**2)) / np.sqrt(2*np.pi*bandwidth**2)
```

```python
x = np.random.uniform(-1, 1, 100)
z = np.zeros(100)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x, kernel(z, x, 0.1), color='blue')
```

```python
y = (x+1)**2 + np.random.randn(x.size)
```

```python
plt.scatter(x, y)
plt.scatter(x, (x+1)**2)
```
