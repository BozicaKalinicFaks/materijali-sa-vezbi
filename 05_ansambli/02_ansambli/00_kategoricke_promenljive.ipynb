```python
import pandas as pd
import numpy as np
```

## Generisemo skup podataka

```python
classes = {
    0: 'cat',
    1: 'mouse',
    2: 'dog',
    3: 'bear',
    4: 'snake',
    5: 'elephant',
}

colors = {
    0: 'brown',
    1: 'white',
    2: 'black',
    3: 'blue',
}

create_classes = np.vectorize(lambda x: classes[x])
create_colors = np.vectorize(lambda x: colors[x])
```

```python
N = 20
xs = np.random.randint(low=0, high=len(classes), size=(N,))
ys = np.random.randint(low=0, high=len(colors), size=(N,))
zs = np.round(np.random.random(size=(N,)), 2)

xs = create_classes(xs)
ys = create_colors(ys)

d = np.array([xs, ys, zs]).T
print(d.shape)
```

```python
df = pd.DataFrame(data=d, columns=['type', 'color', 'age'])
```

## Dummy coding
Dummy coding ili One-Hot encoding su nazivi za enkodiranje koje uzima kategoricku promenljivu sa $k$ mogucih vrednosti
i enkodira je sa $k$ novih promenljivih koje predstavljaju indikatore o originalnoj vrednosti.


Kategoricki atribut `type` kodiramo u dummy coding.

```python
df_dummied = pd.get_dummies(df, columns=['type'])
```

```python
df_dummied
```
