# Pima Indians Diabetes

Skup podataka "Pima Indians Diabetes" se sastoji od 768 instanci od kojih svaka predstavlja vrednosti dobijene analizom pacijenata. Svi atributi su numerički i opseg vrednosti varira od atributa do atributa. Svakoj instanci je pridruzena klasa 1 ili 0 u zavisnosti od toga da li je pacijentu dijagnostifikovan dijabetes u poslednjih 5 godina ili ne.

Naš cilj je da napravimo model koji će na osnovu pomenutih parametara vršiti predikciju da li pacijent boluje od dijabetesa ili ne.

### Korak 1: Uključivanje potrebnih modula

```python
import numpy as np
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

### Korak 2: Učitavanje skupa podataka i podela na trening i test skup

```python
dataset = np.loadtxt('datasets/pima_indians_diabetes.csv', delimiter=',')
```

```python
x = dataset[:, 0:8]
y = dataset[:, 8]
```

```python
(x_train, x_test, y_train, y_test) = train_test_split(x, y, test_size=0.2)
```

```python
print('Skup za treniranje - x: {0}, y: {1}'.format(x_train.shape, y_train.shape))
```

```python
print('Skup za testiranje - x: {0}, y: {1}'.format(x_test.shape, y_test.shape))
```

```python
# standardizacija skupa za treniranje 
mean = np.mean(x_train, axis = 0)
std = np.std(x_train, axis = 0)
x_train = (x_train - mean)/std
```

```python
# standardizacija skupa za testiranje
x_test = x_test - mean
x_test = x_test/std
```

### Korak 3: Pravljenje troslojne neuronske mreže

* Arhitektura mreže:
  * Sloj 1: 15 neurona sa relu aktivacionom funkcijom 
  * Sloj 2: 8 neurona sa relu aktivacionom funkcijom
  * Sloj 3: 1 neuron sa sigmoidnom aktivacionom funkcijom
  
* Funkcija greške: unakrsna entropija (eng. binary cross entropy)
* Optimizator: Adam

* Treniranje mreže:
    * broj epoha: 50
    * veličina paketica za treniranje (eng. batch size): 32

Definisemo arhitekturu mreze.

```python
model = Sequential()
model.add(Dense(15, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs = 50
batch_size = 32
verbose = 0
```

Vrsimo obucavanje mreze.

```python
history = model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, verbose=verbose)
```

```python
acc = history.history['acc']
loss = history.history['loss']
```

```python
plt.xlabel('epochs')
plt.ylabel('acc')
plt.plot(range(epochs), acc, label='training')
plt.legend(loc='best')
```

```python
plt.xlabel('epochs')
plt.ylabel('loss')
plt.plot(range(epochs), loss, c='red', label='training')
plt.legend(loc='best')
```

```python
# evaluacija
train_scores = model.evaluate(x_train,y_train)
test_scores = model.evaluate(x_test, y_test)


# ispis rezultata
print('Skup za treniranje {0}: {1:.4}'.format(model.metrics_names[1], train_scores[1]*100))
print('Skup za testiranje {0}: {1:.4}'.format(model.metrics_names[1], test_scores[1]*100))
```

```python
# dobijanje predikcija za test skup
predictions = model.predict(x_test)
print(x_test.shape)
print(predictions.shape)
```

```python
# npr. predikcija za 13. instancu
index = 12
print('Instanca: {}'.format(x_test[index]))
prediction = 0 if predictions[index]<0.5 else 1
print('Izlaz mreze: {}'.format(predictions[index]))
print('Predikcija: {}'.format(prediction))
print('Tacna vrednost: {}'.format(y_test[12]))
```
