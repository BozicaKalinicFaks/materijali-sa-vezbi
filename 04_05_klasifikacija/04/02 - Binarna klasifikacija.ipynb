# Binarna klasifikacija

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
np.random.seed(10)
```

U zadatku binarne klasifikacije, ciljna promenljiva može imati dve vrednosti. Njih obično obeležavamo sa `0` i `1` ili `-1` i `1`. Za klasu kojoj je pridruženo obeležje `0` (ili `-1`) kažemo da je **negativna** klasa, dok za klasu kojoj je pridruženo obeležje `1` kažemo da je **pozitivna** klasa. 

**Logististička regresija** je jedan od osnovnih modela binarne klasifikacije. Ona pretpostavlja linearnu vezu između ulaznih veličina $X_1$, $X_2$, ..., $X_m$. Ciljna funkcija logističke regresije je zadata formulom $$f_\beta(X)= \frac{1}{1+e^{-(\beta_0 + \beta_1X_1 + \ldots + \beta_mX_m)}} = \frac{1}{1+e^{-\beta^T X}} = \sigma(\beta^TX)$$ gde je $\sigma(x)$ sigmoidna funkcija određena jednačinom $\sigma(x) = \frac{1}{1+e^{-x}}$. Vrednost ciljne funkcije se može interpretirati kao verovatnoća pripadnosti odredjenoj klasi: ukoliko je njena vrednost veća od 0.5, klasifikator predviđa pozitivnu klasu, dok u suprotnom predviđa negativnu klasu. 

## Sigmoidna funkcija

```python
# proizvoljna ekvidistantna mreza 100 tacaka
N = 100
x = np.linspace(-10, 10, N)
y = 1/(1 + np.exp(-x))
```

```python
plt.plot(x, y, label = 'sigmoidna funkcija')
plt.plot(x, [0.5]*N, color = 'red')
plt.plot(x[70], y[70], color='blue', marker='o')
plt.legend(loc = 'best')
plt.show()
```

Ukoliko je vrednost sigmoidne funkcije u tački `x` veća od `0.5`, njoj se pridružuje obeležje `1`. U suprotnom joj se pridružuje obeležje `0`.

## Unakrsna entropija

Parametri modela logističke regresije se određuju minimizacijom funkcije gubitka. U slučaju logističke regresije to je unakrsna entropija određena izrazom $$-\sum_{i=1}^{N} (y_i logf_\beta(x_i) + (1-y_i)log(1-f_\beta(x_i)))$$ u kojem $(x_i, y_i)$ predstavljaju instance skupa za treniranje, a $N$ ukupan broj instanci ovog skupa.

```python
# proizvoljna ekvidistantna mreza 50 tacaka intervala [0, 1]
N = 50
y_i = np.linspace(0, 1, N, endpoint=True)

# vrednosti funkcije log(y)
loss_part_positive = -np.log(y_i) 

# vrednosti funkcije log(1-y)
loss_part_negative = -np.log(1-y_i)
```

```python
plt.figure(figsize = (10, 5))

plt.subplot(1, 2, 1)
plt.title('Doprinos funkciji gubitka pozitivnih instanci')
plt.plot(y_i, loss_part_positive, color ='b')

plt.subplot(1, 2, 2)
plt.title('Doprinos funkciji gubitka negativnih instancu')
plt.plot(y_i, loss_part_negative, color ='r')

plt.show()
```

Ako je za instancu $(x_i, y_i)$ očekivana vrednost klasifikatora $1$, što više grešimo sa predikcijom $y_i$ i što više odstupamo od očekivane vrednosti, veći je i doprinos funkciji gubitka (grafik levo). Ako je očekivana vrednost klasifikatora $0$, što više odstupamo od ove vrednosti, veći je doprinos funkciji gubitka (grafik desno). 

# Primer

U primeru koji sledi model logističke regresije biće iskorišćen za klasifikaciji tumora na maligne i benigne. Koristićemo Viskonsis skup podataka o kojem se može više pročitati na zvaničnom [UCI](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)) sajtu.

```python
from sklearn import linear_model
from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
```

### Analiza podataka

```python
# print(data.DESCR)
```

Svaka instanca skupa opisana je sa 30 različitih atributa.

```python
data.feature_names
```

```python
X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
X.shape
```

```python
# X.info()
```

Vrednosti ciljne promenljive su `0` i `1` i redom su vezane za maligne i benigne tumore.

```python
data.target_names
```

```python
y = data.target
```

```python
print('Broj benignih instanci: ')
np.sum(y==1)
```

```python
print('Broj malignih instanci: ')
np.sum(y==0)
```

### Podela podataka na skup za treniranje i skup za testiranje

Prilikom podele skupa podataka na skupove za treniranje i testiranje, vodićemo računa o `stratifikaciji` (parametar stratify). Stratifikacija je način podele podataka kojim se čuva distribucija klasa. 

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.33, random_state = 7, stratify = y)
```

```python
X_test.shape
```

Provera stratifikovanosti:

```python
m_train = np.sum(y_train == 0)
b_train = np.sum(y_train == 1)
print('Benigni: ', b_train, 'Maligni: ', m_train)
```

```python
m_test = np.sum(y_test == 0)
b_test = np.sum(y_test == 1)
print('Benigni: ', b_test, 'Maligni: ', m_test)
```

```python
plt.title("Vizuelna provera stratifikovanosti")
plt.xticks([0,1])
plt.xlabel(' 0 - maligni tumor, 1 - benigni tumor')
plt.hist([y_train, y_test], color=['orange', 'cadetblue'], label=['skup za treniranje', 'skup za testiranje'])
plt.legend(loc='best')
plt.show()
```

### Standaradizacija podataka

```python
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Učenje modela

```python
model = linear_model.LogisticRegression(solver='lbfgs')
```

```python
model.fit(X_train, y_train)
```

Rezultujući parametri se mogu pročitati kroz `intercept_` i `coef_` svojstva.

```python
model.intercept_
```

```python
model.coef_
```

Interpretacija vrednosti koeficijenata je važna zbog razumevanja samog modela i uticaja atributa.

```python
N = len(data.feature_names)
values = model.coef_[0]
plt.figure(figsize = (10, 5))
plt.bar(np.arange(0, N),values)
plt.xticks(np.arange(0, N), data.feature_names, rotation='vertical')

plt.show()
```

Na primer, na ovaj način se može uvideti da atributi kao što su `radius error`, `worst texture` ili `worst area` važni u predviđanju negativne klase tj. maligni tumor, doku je atribut `compactness error` važan za predikciju pozitivne klase tj. benigni tumor. 

### Evaluacija

```python
y_test_predicted = model.predict(X_test)
```

Sve mere od interesa se nalaze u `metrics` biblioteci.

```python
metrics.accuracy_score(y_test, y_test_predicted)
```

```python
metrics.precision_score(y_test, y_test_predicted)
```

```python
metrics.recall_score(y_test, y_test_predicted)
```

```python
metrics.f1_score(y_test, y_test_predicted)
```

```python
y_train_predicted = model.predict(X_train)
```

```python
train_score = metrics.accuracy_score(y_train, y_train_predicted)
```

```python
test_score = metrics.accuracy_score(y_test, y_test_predicted)
```

```python
print("Tacnost na skupu za treniranje: {train}\nTacnost na skupu za testiranje: {test}".format(train=train_score, test=test_score))
```

Sumarna evaluacija po klasama može se pročitati iz klasifikacionog izveštaja dostupnog u `metrics` biblioteci.

```python
print(metrics.classification_report(y_test, y_test_predicted))
```

`Makro prosek` (macro avg) se računa kao aritmetička sredina dobijenih vrednosti po klasama. Na primer, makro prosek za F1-meru se računa kao $\frac{0.97+0.98}{2} = 0.975 $

`Težinski prosek` uzima u obzir i broj instanci koje služe kao potpora ocenjivanju. Na primer, težinski prosek za preciznost se računa kao $\frac{70}{188} \cdot 0.96 + \frac{118}{188} \cdot 0.99 = 0.978829$

Matrica konfuzije se može dobiti pozivom metode `confusion_matrix` biblioteke `metrics`.

```python
metrics.confusion_matrix(y_test, y_test_predicted)
```

Organizacija ove matrice je sledeća: 
<br> 
na poziciji (0, 0) se nalazi TN vrednost, na poziciji (1, 0) FN vrednost, na poziciji (1, 1) je TP vrednost, i konačno, na poziciji (0, 1) je FP vrednost. 

Za detaljniju analizu prstora greški je često potrebno proučiti instance obeležene kao lažno pozitivne (FP) ili lažno negativne (FN). 

```python
# Izdvajanje lažno pozitivnih (FP) instanci
# False positive: y_true = 0, y_predicted = 1

FP_mask = (y_test == 0) & (y_test_predicted == 1)
```

```python
np.where(FP_mask == True)
```

```python
X_test[38]
```

```python
# Izdvajanje lažno negativnih (FN) instanci 
# False negative: y_true =1, y_predicted = 0
FN_maska = (y_test == 1) & (y_test_predicted == 0)
```

```python
# broj lazno negativnih instanci
np.sum(FN_maska)
```

```python
# indeksi lazno negativnih instanci
ind = np.where(FN_maska == True)
ind
```

```python
# lazno negativne instance
X_test[ind]
```

### Analiza sigurnosti klasifikatora

Biblioteka omogućava da se pomoću funkcije modela `predict_proba` za svaku instancu pojedinačno dobiju i ocene verovatnoća pripadnosti pozitivnoj tj. negativnoj klasi. 

Na primer, za prvih 10 instanci klasifikator daje sledeći rezultat: 

```python
y_test_predicted[0:10]
```

Verovatnoće pripadnosti negativnoj tj. pozitivnoj klasi za ovih 10 instanci se mogu dobiti sledećim kodom:

```python
y_probabilities_predicted = model.predict_proba(X_test)
y_probabilities_predicted[0:10]
```

Oko predikcije nekih vrednosti klasifikator može biti vrlo siguran, dok u nekim slučajevima to ne mora biti tako. Postavljanjem odgovarajućeg praga može se vršiti prilagođavanje problemu koji se rešava i uticati na strogost klasifikatora.

```python
model_confidence = []
```

```python
for p in y_probabilities_predicted: 
    model_confidence.append(np.abs(p[0]-p[1]))
```

```python
model_confidence = np.array(model_confidence)
```

```python
# instance uredjene na osnovu sigurnosti klasifikatora u svoju odluku
# prvo su navedeni indeksi instanci oko cije klasifikacije klasifikator nije sasvim siguran
model_confidence.argsort()
```

```python
model_confidence.sort()
```

```python
# uredjeni niz razlike verovatnoca 
model_confidence
```

Slična funkcionalnost je dostupna i preko metode `predict_log_proba` modela. Jedina razlika je u tome što su verovatnoće pripadanja na logaritamskoj skali. 

```python
y_log_probabilities_predicted = model.predict_log_proba(X_test)
y_log_probabilities_predicted[0:10]
```
